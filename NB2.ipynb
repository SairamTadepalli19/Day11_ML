```python
#Import the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import the library

data = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
#Seperate x and y variables

x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
#Split the dataset into both training and testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.20, random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
#We need to implement feature scaling (Standard Scaler Operation) (Both positives and negatives will be considered)

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
#Transform the x_train and fit , transform x_test

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
#Implement Gaussian NB

from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()

classifier.fit(x_train, y_train)
```

```python
#Test with the new results

print(classifier.predict(sc.transform([[20, 87000]])))
```

```python
#Predict the test set results

y_pred = classifier.predict(x_test)
```

```python
#Confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```

```python
#Visualize the training set results

from matplotlib.colors import ListedColormap

x_set, y_set = sc.inverse_transform(x_train), y_train

x1,x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 0, stop = x_set[:, 0].max() + 0 ),
                    np.arange(start = x_set[:, 1].min() - 0, stop = x_set[:, 1].max() + 0))

plt.contourf(x1,x2, classifier.predict(sc.transform(np.array([x1.ravel(),x2.ravel()]).T)).reshape(x1.shape),
             cmap = ListedColormap(('green', 'red')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())


for i,j in enumerate(np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j,1], c = ListedColormap(("blue", "yellow"))(i), label = j)

plt.title("NB Classifier")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```
